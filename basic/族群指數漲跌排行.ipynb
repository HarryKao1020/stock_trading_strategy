```python
!pip install shioaji
import numpy as np
import pandas as pd
import requests
import shioaji as sj
import datetime
from datetime import date, timedelta
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from google.colab import userdata



# 建立API物件，simulation=True是代表測試帳號
api = sj.Shioaji(simulation=True)



#==== for Colab
shioaji_secret=userdata.get('SHIOAJI_SECRETKEY')
shioaji_apikey=userdata.get('SHIOAJI_APIKEY')
accounts = api.login(shioaji_apikey, shioaji_secret)
```

```python
print(api.Contracts.Indexs.TSE)
print(api.Contracts.Indexs.OTC)
```

```python
contracts_dict = []
for i in range(4,44):
    contracts = api.Contracts.Indexs.TSE[f'{i:03d}']
    if(contracts==None):
        continue
    else:
      contracts_dict.append(contracts)



contract = contracts_dict
snapshots = api.snapshots(contract)
# df = pd.DataFrame(s.__dict__ for s in snapshots)
df = pd.DataFrame(
    {
        **{key: [getattr(s, key) for s in snapshots] for key in snapshots[0].__dict__.keys()},
        "name": [contract.name for contract in contracts_dict]
    }
)
df.ts = pd.to_datetime(df.ts)

# 根據 'change_rate' 列由大到小排序
df_sorted = df.sort_values(by='change_rate', ascending=False)

# 顯示排序後的 DataFrame
df_sorted[['name','change_rate']]

```

```python
otc_contracts_dict = []
for i in range(102,167):
    otc_contracts = api.Contracts.Indexs.OTC[f'{i:03d}']
    if(otc_contracts==None):
        continue
    else:
      otc_contracts_dict.append(otc_contracts)



otc_contract = otc_contracts_dict
snapshots = api.snapshots(otc_contract)
# df = pd.DataFrame(s.__dict__ for s in snapshots)
otc_df = pd.DataFrame(
    {
        **{key: [getattr(s, key) for s in snapshots] for key in snapshots[0].__dict__.keys()},
        "name": [otc_contract.name for otc_contract in otc_contracts_dict]
    }
)
otc_df.ts = pd.to_datetime(otc_df.ts)
# 根據 'change_rate' 列由大到小排序
otc_df_sorted = otc_df.sort_values(by='change_rate', ascending=False)

# 顯示排序後的 DataFrame
otc_df_sorted[['name','change_rate']]
```
